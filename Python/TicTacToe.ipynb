```python
import random

class TicTacToe:

  def __init__(self):
    self.board = []  #The board is a list, filled with  lists(each row is represented as a list) -> internal board representation is a list of lists

  #Function to create the board.
  #Creates a row as a lists, and does so 3 times.
  #At the start of the game each field is represented as a _.
  def create_board(self):
    for i in range(3):
      row = []  
      for j in range(3):
        row.append('_') 
      self.board.append(row)

  #Function to print the board for each round.
  def show_board_to_user(self):
    for row in self.board:
      for column in row:
        print(column, end=" ") #By default print() ends with a new line, with the end=" " it makes the print end with a space. Now every item row is print as one row, instead of a row for each item.
      print()

  #Function to print out the rules.
  def explain_game(self):
    print('The user is represented as an "X" and the cpu as an "O".\n'
    + 'The row and column count starts at zero and ends at two. \n'
    + 'First input is the row, followed by the column.\n'
    + 'Seperate ur input with a space and press enter after.\n'
    + 'As an example the input: 0 0 would result in marking the upper left corner.\n'
    + 'Visualization of the input  "0 0" :\n'
    + 'X _ _ \n'
    + '_ _ _ \n'
    + '_ _ _ \n'
    + 'Each field can only be marked once.\n'
    + 'To win the game either player has to mark three fields in the same row, column or diagonal.\n'
    + 'If the game is a draw each field has to be marked to end the game.\n')

  #Function to mark the user fields.
  def mark_field_user(self, row, column):
    self.board[row][column] = 'X'

  #Function to mark the cpu fields.
  def mark_field_cpu(self, row, column):
    self.board[row][column]  = 'O'

  #Tests if the user input is valid.
  def is_input_valid(self, row, column):
    if self.board[row][column] == '_':
      return True
    else:
      return False
  
  #Function to check if the board is filled, to handle the draw scenario.
  def is_filled(self):
    for row in self.board:
      for column in row:
        if column == '_':
          return False
    return True

  #Function to check if a field is still empty. Used for the user input and for cpu logic.
  def is_empty_field(self,row, column):
    if self.board[row] [column] == '_':
      return True
    return False

  #Function for a  random cpu move
  #If the cpu can't win and hasn't to stop the human from winning it will perform a random move
  def cpu_random(self):
    while True: #So with while True, the follwing code will run forever until i put a break. I use it so there is going to be a marked field from the cpu.
      row = random.randint(0, 2)
      column = random.randint(0, 2) 
      if self.is_empty_field(row, column): 
        self.mark_field_cpu(row, column) 
        break

  #Checks if the human player could win this turn and marks the needed fiel to stop him from doing so.
  #If their is a free spot and already 2 symbols of the human, it will mark the field
  def stop_user_from_winning(self):
    length_of_board = len(self.board)
    #First check the rows
    for row in range(length_of_board):
      freeFieldCount = 0
      symboleCount = 0
      for column in range(length_of_board):
        if self.is_empty_field(row, column): #Checks if their is an empty field in the row.
          freeFieldCount += 1
        if self.board[row][column] == 'X': #Checks if their is already a human marked field in the row
          symboleCount += 1 
      if freeFieldCount > 0 and symboleCount == 2: 
        for column in range(length_of_board): #Iterrates over all columns of that row and spots the free field and marks it after.
          if self.is_empty_field(row, column):
            self.mark_field_cpu(row, column)
            return True

    #Check the columns
    #If their is a free spot and already 2 symbols of the human, it will mark the field
    for column in range(length_of_board):
      freeFieldCount = 0
      symboleCount = 0
      for row in range(length_of_board):
        if self.is_empty_field(row, column): #Checks if their is an empty field in the row.
          freeFieldCount += 1
        if self.board[row][column] == 'X': #Checks if their is already a human marked field in the row
          symboleCount += 1 
      if freeFieldCount > 0 and symboleCount == 2: 
        for row in range(length_of_board): #Iterrates over all rows of that column and spots the free field and marks it after.
          if self.is_empty_field(row, column):
            self.mark_field_cpu(row, column)
            return True
    
    #Check first diagonal
    #Checks if their are already 2 marked fields in the first diagonal and then marks the free field.
    symboleCount = 0
    freeFieldCount = 0
    for pair in range(length_of_board): #Itterates over all pairs ((0,0). (1,1), (2,2)).
      if self.is_empty_field(pair, pair): #Checks if their is a free field in the first diagonal
        freeFieldCount += 1
      if self.board[pair][pair] == 'X': #Checks if their is a human symbol.
        symboleCount += 1
      if freeFieldCount > 0 and symboleCount == 2: 
        for pair in range(length_of_board):
          if self.is_empty_field(pair, pair):
            self.mark_field_cpu(pair, pair)
            return True

    #Check second digonal
    #Checks if their are already 2 marked fields and a freee field in the second diagonal, marks the field after.
    symboleCount = 0  
    freeFieldCount = 0
    for diag in range(length_of_board): #Iterrsates over the second diagonal ((0,2), (1,1), (2,0))
      if self.is_empty_field(diag, length_of_board - 1 - diag): #Checks if their is a free field in the second diagonal.
        freeFieldCount += 1
      if self.board[diag][length_of_board - 1 - diag] == 'X': #Checks if their is a human symbol.
        symboleCount += 1
      if freeFieldCount > 0 and symboleCount == 2: 
        for diag in range(length_of_board):
          if self.is_empty_field(diag, (length_of_board - 1 - diag)):
            self.mark_field_cpu(diag, (length_of_board - 1 - diag))
            return True
    return False

  #Function to check if the cpu can mark a field to win the game, and will mark the field if possible.
  #If their is a free spot and already 2 symbols of the cpu, it will mark the field
  def win_game_for_cpu(self):
    length_of_board = len(self.board)
    #Check rows
    for row in range(length_of_board):
      freeFieldCount = 0
      symboleCount = 0
      for column in range(length_of_board):
        if self.is_empty_field(row, column): #Checks if their is an empty field in the row.
          freeFieldCount += 1
        if self.board[row][column] == 'O': #Checks if their is already a cpu marked field in the row
          symboleCount += 1 
      if freeFieldCount > 0 and symboleCount == 2: 
        for column in range(length_of_board): #Iterrates over all columns of that row and spots the free field and marks it after.
          if self.is_empty_field(row, column):
            self.mark_field_cpu(row, column)
            return True
    
    #Check Columns
    #If their is a free spot and already 2 symbols of the cpu, it will mark the field
    for column in range(length_of_board):
      freeFieldCount = 0
      symboleCount = 0
      for row in range(length_of_board):
        if self.is_empty_field(row, column): #Checks if their is an empty field in the row.
          freeFieldCount += 1
        if self.board[row][column] == 'O': #Checks if their is already a cpu marked field in the row
          symboleCount += 1 
      if freeFieldCount > 0 and symboleCount == 2: 
        for row in range(length_of_board): #Iterrates over all rows of that column and spots the free field and marks it after.
          if self.is_empty_field(row, column):
            self.mark_field_cpu(row, column)
            return True

    #Check first diagonal
    #Checks if their are already 2 marked fields in the first diagonal and then marks the free field.
    symboleCount = 0
    freeFieldCount = 0
    for pair in range(length_of_board): #Itterates over all pairs ((0,0). (1,1), (2,2)).
      if self.is_empty_field(pair, pair): #Checks if their is a free field in the first diagonal
        freeFieldCount += 1
      if self.board[pair][pair] == 'O': #Checks if their is a cpu symbol.
        symboleCount += 1
      if freeFieldCount > 0 and symboleCount == 2: 
        for pair in range(length_of_board):
          if self.is_empty_field(pair, pair):
            self.mark_field_cpu(pair, pair) 
            return True 
      
    #Check second diagonal
    #Checks if their are already 2 marked fields and a freee field in the second diagonal, marks the field after.
    symboleCount = 0  
    freeFieldCount = 0
    for diag in range(length_of_board): #Iterrsates over the second diagonal ((0,2), (1,1), (2,0))
      if self.is_empty_field(diag, length_of_board - 1 - diag): #Checks if their is a free field in the second diagonal.
        freeFieldCount += 1
      if self.board[diag][length_of_board - 1 - diag] == 'O': #Checks if their is a cpu symbol.
        symboleCount += 1
      if freeFieldCount > 0 and symboleCount == 2: 
        for diag in range(length_of_board):
          if self.is_empty_field(diag, (length_of_board - 1 - diag)):
            self.mark_field_cpu(diag, (length_of_board - 1 - diag))
            return True
    return False



  #Function for the cpu turn.
  def cpu_turn(self):
    length_of_board = len(self.board)
    turn = 0

    #First check if the next move could win the game. for the cpu.
    if turn == 0:
      if self.win_game_for_cpu():
       turn += 1

    #Now check if human needs to be stopped from winning
    if turn == 0:
      if self.stop_user_from_winning():
       turn += 1

    #Else just do a random move.
    if turn == 0:
      self.cpu_random()
      turn += 1

    
  #Function to check for a human win.

  def human_win(self):
    length_of_board = len(self.board)
    
    #First check rows
    symboleCount = 0
    for row in range(length_of_board): #Goes trough all rows, if the user symbol appears 3 times it returns True
      symboleCount = 0
      for column in range(length_of_board):
        if self.board[row][column] == 'X':
          symboleCount += 1
      if symboleCount == 3:
        return True
    
    #Check columns
    symboleCount = 0
    for column in range(length_of_board): #Goes trough all columns, if the user symbol appears 3 times it returns True.
      symboleCount = 0
      for row in range(length_of_board):
        if self.board[row][column] == 'X':
          symboleCount += 1
      if symboleCount == 3:
        return True
    
    #Check first diagonal
    symboleCount = 0
    for pair in range(length_of_board): #Goes trough the first diagonal, if the user symbol appears 3 times, it returns True.
      if self.board[pair][pair] == 'X':
        symboleCount += 1
    if symboleCount == 3:
      return True
    
    #Check second diagonale
    symboleCount = 0
    for diag in range(length_of_board): #Goes trough the second diagonal, if the user symbol appears 3 times, it returns True
      if self.board[diag][length_of_board - 1 - diag] == 'X':
        symboleCount += 1
    if symboleCount == 3:
      return True
    return False #If there isnt a win condition, the function returns False

  #Function to check for a cpu win
  def cpu_win(self):
    length_of_board = len(self.board)
    
    #First check rows
    symboleCount = 0
    for row in range(length_of_board): #Goes trough all rows, if the cpu symbol appears 3 times it returns True
      symboleCount = 0
      for column in range(length_of_board):
        if self.board[row][column] == 'O':
          symboleCount += 1
      if symboleCount == 3:
        return True
    
    #Check columns
    symboleCount = 0
    for column in range(length_of_board): #Goes trough all columns, if the cpu symbol appears 3 times it returns True.
      symboleCount = 0
      for row in range(length_of_board):
        if self.board[row][column] == 'O':
          symboleCount += 1
      if symboleCount == 3:
        return True
    
    #Check first diagonal
    symboleCount = 0
    for pair in range(length_of_board): #Goes trough the first diagonal, if the cpu symbol appears 3 times, it returns True.
      if self.board[pair][pair] == 'O':
        symboleCount += 1
    if symboleCount == 3:
      return True
    
    #Check second diagonale
    symboleCount = 0
    for diag in range(length_of_board): #Goes trough the second diagonal, if the cpu symbol appears 3 times, it returns True
      if self.board[diag][length_of_board - 1 - diag] == 'O':
        symboleCount += 1
    if symboleCount == 3:
      return True
    return False #If there isnt a win condition, the function returns False

  #Function to start the game.
  def start_game(self):
    self.create_board() #Creates the board.
    self.explain_game() #Shows the rules to the user.
    print("User's turn: \n")
    self.show_board_to_user()
    while True: #While True is useed, to keep the game running until either human or cpu win or the game ends as a draw
      print('')
      row, column = list(map(int, input("Please enter row and column numbers u want to mark: ").split())) #Read in the user and safe it as a integer list
      if self.is_empty_field(row, column): #check if the field is free, else ask for a new input
        self.mark_field_user(row, column)
      else:
        while True:
          row, column = list(map(int, input('This field is already marked, please choose a new one: ').split()))
          self.mark_field_user(row, column)
          break
      self.show_board_to_user()
      if self.human_win(): #check if u human player did win after his turn
        print('Human player wins!')
        break
      if self.is_filled(): #check if its a draw
        print("It's a draw!")
        break
      print('')
      print("CPU's turn:\n")
      self.cpu_turn() #process the cpu turn
      self.show_board_to_user()
      if self.cpu_win(): #check for cpu win
        print("The CPU wins!")
        break
      if self.is_filled(): #check for draw again
        print("It's a draw!")
        break
      print('')
      print("User's turn: \n")

tictactoe = TicTacToe()
tictactoe.start_game()
```
